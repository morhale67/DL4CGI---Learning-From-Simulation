```python
import pickle
import os
from pathlib import Path
from typing import Dict, Any

def obtain_res_dict(pkl_file_path: str) -> Dict[str, Any]:
    
    with open(pkl_file_path, "rb") as f:
        return pickle.load(f)
    

def analyze_results(results_dir_path: str):
    instances_dirs = os.listdir(Path(results_dir_path))
    
    for execution_instance in instances_dirs:
        d = obtain_res_dict(
            Path(
                results_dir_path,
                execution_instance,
                "numerical_outputs.pkl"
            )
        )
        
        best_epoch = d["train_loss"].index(min(d["train_loss"]))
        from pprint import pprint
        pprint(d)
        print()
```

```python
analyze_results("Results/")
```
